# Linear Regression

----
linear regression is one of the most widely used regression models. The model seek to identify a linear relationships between the dependent and independent variable(s). It models the expected value of a quantitative response (dependent) variable  from one or more explanatory variables (independent variables). 



# Simple Linear Regression and ordinary least squares (OLS)
----

simple linear regression involves predicting a quantitative response variable from a single explanatory variable


Let consider a simple linear regression to determine how the number of flyers sold on a day relate or determine the Sales of the day defined by
$$Y_{i} = \beta_{0} +\beta_{1}X_{i} + \epsilon _{i}$$ 
where 
<ui>
    <li>$\epsilon _{i}$ are assumed to be independent random variables having a normal distribution with mean zero and constant
variance $\sigma^{2}$ </li>
    <li>$Y_{i}$=Sales for the ith day</li>
    <li>$X_{i}$=number of flyers sold on the ith day</li>
    <li>$\beta_{0} $=the intercept of the line (the predicted amount of sales when no flyers is sold)</li>
    <li>$\beta_{1}$=the slope of the line (the slope represent the change in Sales (Y) for a unit change in the number of flyers (X) sold))</li>
   </ui>
   
  Using Simple Linear Regression, the model assumes there is approximately a linear relationship between the number of flyers (X) sold on a day and the Sales(Y) for that day  defined by

$$Y \approx \beta_{0} +\beta_{1}X $$ 

 
 # NOTE
 The intercept and slope are unknown constant of the model known as the model coefficients or parameters. 

In order to be able to predict the expected amount of Sales for the day based or conditioned on the number of flyers sold, the unknown parameters $ \beta_{0}$ and $\beta_{1}$ are estimated using training data as $\hat \beta_{0}$ and $\hat \beta_{1}$.
such that
$$\hat y_{i}=\hat \beta_{0} +\hat \beta_{1}X_{i}$$
where $\hat y_{i} $ is the  expected  predicted Sales (Y) of the day conditioned on the number of flyers (the ith value of X ) sold on that day using the estimated parameters.

# NOTE
Using the model our goal is to predict the expected amount of Sales (y) for a day based or conditioned on the number of flyers sold defined as

$$E(Y)\approx \hat \beta_{0} +\hat \beta_{1}X_{i}$$



$\hat \beta_{0}$  and $\hat \beta_{1}$ are then estimated using training data in such a way that minimize the residual sum of squares (RSS or $SS_{res}$) errors  defined as
$$RSS=SS_{res}=e^{2}=\sum_{i}^{n}(y_{i}-\hat y_{i})^{2}$$

$ the \ residual=e_{i} = y_{i} âˆ’ \hat y_{i}$ represents the difference between the ith observed (True value of y) response value and the predicted value for the ith observation by the model

The estimates are defined as
$$ \hat \beta_{0}= \bar y -\hat \beta_{1} \bar x$$

$$\hat \beta_{1}=\frac{\sum_{i}^{n}(x_{i}-\bar x) (y_{i}-\bar y)}{\sum_{i}^{n}(x_{i}-\bar x)^{2}}= \frac{ \bar y \sum_{i}^{n}x_{i}-\sum_{i}^{n}x_{i}y_{i}}{\bar x \sum_{i}^{n}x_{i}-\sum_{i}^{n}x_{i}^{2}}$$

where $$\hat y=\sum_{i}^{n}\frac{y_{i}}{n}$$
where $$\hat x=\sum_{i}^{n}\frac{x_{i}}{n}$$


# variance of the residual ( variability within residuals)
## $$Var(e)=\frac{1}{n-2} \sum_{i}^{n} e _{i}^{2} $$

# residual standard error   
The residual standard error  of the model, denoted by SE, is our estimate of the standard deviation of the noise in Y (Sales) (the variation in it that is considered unexplainable).

## $$SE=\sqrt{\frac{1}{n-2} \sum_{i}^{n} e _{i}^{2}  } $$

# NOTE
## $$ \sigma ^{2}=Var(e)$$

## $$SE(\hat \beta_{1})^{2}= \left(  \frac{\sigma^{2} }{\sum_{i}^{n}(x_{i}-\bar x)^{2}}  \right)  $$




# HYPOTHESIS
----
Researchers usually collect statistical evidence before making a claim, and this is done by conducting a test of statistical significance. The first step in conducting a test of statistical significance is to state the hypothesis (the hypothesis to be tested and is usually called the <b>Null Hypothesis</b>), collect or sample enough data from the population and use a hypothesis-testing technique to test whether the evidence (estimates from the sampled data) provided by data about the claim should be rejected or accepted. A statistical hypothesis is an assumption (or guess) about unknown properties of some measurement about populations and Hypothesis-testing provides the techniques for determining whether a statistical hypotheses should be accepted or rejected (refuted). Using inferential statistics we then draw conclusion about populations based on the estimates. 


# T-TEST

----
 Standard errors can also be used to perform hypothesis tests on the hypothesis coefficients. The most common hypothesis test involves testing 
 

 <b>the NULL HYPOTHESIS $H_{O}$</b> defined as

$$H_{O} : \ There  \ is \ no \ relationship \ between \ X  \ and \ Y $$ denoted as
$$ H_{O}: \beta_{1}=0 $$

 By assuming $H_{O} : \ There  \ is \ no \ relationship \ between \ X  \ and \ Y $ implies that any relationship/association seen between X and Y must be entirely due to a random chance.

# Alternative hypothesis

How do we know that the relationship between X and Y is real and not just random chance due to random sampling error? We need to find a way to test the null hypothesis and see if we can reject this possibility. To test the null hypothesis we state an alternative hypothesis ( what we are trying to prove to be true)
$$H_{a} :\  There \ is \ some \ relationship \ between \ X \ and \ Y$$
denoted as
$$ H_{a}: \beta_{1}\neq 0 $$
 
 if the is no relationship between X and Y then $ \beta_{1}=0$ and the model reduces to $Y=\beta_{0} +\epsilon$

Using t-statistic test we want to test whether we should accept or reject $H_{O}$ and is defined by
$$t=\frac{\hat \beta_{1}-0}{SE(\hat \beta_{1})} $$



# P-VALUE
---
To test if the null hypothesis is true,  we simply ask, how often could we get a result as big as we did if the null were true and the answer to that question is called a p-value and therfore

-   *p*-values represent at what percentage can the relationship be as a result of some randon  chance
-  If the *p*-value for a result or relationship is small then it means it would be rare to get that
    result by a random chance 
-   If the *p*-value for a result or relationship is large then it means that relationship is resulting from random chance
-  A small *p*-value (typically, less than 5% or "< . 05") tells us that the data and null are highly incompatible and that there relationship between X and Y is not due some randon chance and we conclude the null hypothesis is false

# $R^{2}$ Coefficient of determination or measures of fit or goodness fit

is the proportion of the variance in the dependent variable that is predictable or accounted or explained by the independent variable(s). $R^{2}$ normally ranges from 0 to 1 (NB can be negative)  with best possible score being 1.0 

and is defined as
$$  R^{2}=\frac{Explained \ Variation}{Total \ Variation }=\frac{SS_{reg}}{ SS_{tot}}=\frac{\sum_{i}^{n}(\hat y_{i}- \bar y)^{2}}{\sum_{i}^{n}(y_{i}- \bar y)^{2}}$$


$$R^{2}=\frac{SS_{tot}-SS_{res}}{ SS_{tot}}=1-\frac{SS_{res}}{ SS_{tot}} $$
$$OR$$
$$R^{2}=\frac{SS_{reg}}{ SS_{tot}}=\frac{SS_{reg}/n}{ SS_{tot} / n} $$ 

# The adjusted R-squared
The adjusted R-squared is a modified version of R-squared for the number of predictors in a model. The adjusted R-squared increases only if the added independent variable improves the model and decreases if the added independent variable improves the model by less than expected by chance. 

$$R^{2} \ Adjusted= 1- \frac{ \frac{SS_{res}}{n-p-1}}{ \frac{SS_{tot}}{n-1}  } $$



$$R^{2} \ Adjusted=1-\frac{(1- R^{2})(n-1)}{n-p-1}  $$

where is p is total number of independent (explanatory) variables in the model (not including the constant term or bias), and n is the sample size

# difference between R-squared and the adjusted R-squared
----
One major difference between R-squared and the adjusted R-squared is that <b>R-squared</b> gives the proportion of variance or percentage of variation in the dependent variable that is accounted for or explained by the independent variables as if all the independent variables in the model really affect the dependent variable. Every independent variable added to a model increases the R-squared even if those variables are only weakly associated with the response and never decreases it. An independent variable that lead to just a tiny increase in $R^{2}$ when added to the model can be dropped because this variable does not have any real improvement in the fitted model.


while the <b>adjusted R-squared </b>gives the gives the proportion of variance or percentage of variation explained only the independent variables that really affect the dependent variable, usually lower than the R-squared. The adjusted R2 increases or decreases when an independent variable is added to the model showing how that independent variable affect the dependent variable and is best suited for multiple linear regression.

# overall F-statistic for regression 
---
The F-test for linear regression tests whether any of the d independent variables in a multiple linear regression model are significant

 In the multiple regression setting with d independent variables or predictors, we need to test whether all of the regression coefficients are zero, 
 
 $$H_{O}:\beta_{0}=\beta_{1}=\cdots= \beta_{d}=0$$.
 $$H_{O}:there \ is \ no \ relationship \ between \ the \ variables$$.
 
versus the alternative
$$H_{a}: \beta_{i} \neq 0 \ for \ at \ least \ one \ value of i  $$

$$F=\frac{ \frac{TSS-SS_{res}}{d}}{\frac{SS_{res}}{n-d-1}}=\frac{SS_{reg}/d}{SS_{res}/ (n-d-1)} $$

 when there is no relationship between the response and the predictors, the F-statistic is expected to take on a value close to 1 (We accept $H_{0}$ when the computed F-value is close to 1)

## To compute the standard errors associated with $\beta _{0}$and $\beta _{1}$
## $$SE(\hat \beta_{0})^{2}= \sigma^{2} \left( \frac{1}{n} + \frac{\bar x}{\sum_{i}^{n}(x_{i}-\bar x)^{2}}  \right)  $$

# residual standard error   


## $$SE=\sigma^{2}=\sqrt{\frac{1}{n-2} \sum_{i}^{n} e _{i}^{2}  } $$



```r
SE_B0=function(x,res){
    SE=sum((res-mean(res))**2/(length(res)-2))
    residu_er=SE   
    b_s=(1/length(x)) +(mean(x)**2/sum((x-mean(x))**2))
    result=sqrt(residu_er*b_s)
    return(round(result,5))
}
```


## $$SE(\hat \beta_{1})^{2}= \left(  \frac{\sigma^{2} }{\sum_{i}^{n}(x_{i}-\bar x)^{2}}  \right)  $$

```r
SE_B1=function(x,res){
    SE=sum((res-mean(res))**2/(length(res)-2))
    residu_er=SE   
    b_s=sum((x-mean(x))**2)
    result=sqrt(residu_er/b_s)
    return(round(result,5))
}
```

```r
SE_B0(x = train$Flyers,res = lrg$residuals)
```

```r
SE_B1(x = train$Flyers,res = lrg$residuals)
```

# Multiple regression

----- 

Multiple regression is an extension of the principles of simple linear regression to situations in which there are two or more explanatory variables.


Let consider a multiple linear regression to determine how the number of independent varaibles in the dataset help in determing the sales of the day defined by
$$Y = \beta_{0} +\beta_{1}x_{1} +\beta_{2}x_{2}+\beta_{3}x_{3}+ \cdots+ \epsilon$$ 
which can be represented compactly using a dot product:

$$Y= \beta X + b$$

we need to estimate the parameters $\hat \beta$ such that
$$\hat y=\hat \beta X$$.

$$\hat y=X\hat \beta=X(X^{T}X)^{-1}X^{T}y $$.

where 
<ui>
    <li>Îµ is a random variable with mean zero </li>
    <li>y=Sales</li>
    <li>X=explanatory variables( Day, Temperature, Rainfall, Flyers, Price,Sales)</li>
    <li>$\beta $=parameters of the model</li>
   </ui>
   
   
   $$\hat \beta =(X^{T}X)^{-1}X^{T}y$$
